# Newton Raphson

## Algorithm: Newton-Raphson Method

1. Define the function `f(x)` and its derivative `f'(x)`.

2. Choose an initial guess `x0`, tolerance `ε`, and maximum iterations.

3. Repeat until `|x1 - x0| < ε`:
   - Compute next approximation:  
     $$x_{1} = x_{0} - \frac{f(x_{0})}{f'(x_{0})}$$
   - Update value:  
     $$x_{0} = x_{1}$$

4. After the loop ends, approximate the root as `x1`.

5. Plot:  
   - Curve of `f(x)`  
   - Red dots at each approximation (`x1` values)  
   - A vertical green dashed line at the final root  

6. Print the final root value.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define function and its derivative
def f(x):
    return x**3 - 5*x - 9   # Example function

def df(x):
    return 3*x**2 - 5       # Derivative of f(x)

# Newton-Raphson Method
def newton_raphson(f, df, x0, tol, max_iter=100):
    roots = []
    for _ in range(max_iter):
        x1 = x0 - f(x0) / df(x0)
        roots.append(x1)
        if abs(x1 - x0) < tol:
            return x1, roots
        x0 = x1
    return x0, roots  # fallback in case it doesn't converge

# Example usage
root, roots = newton_raphson(f, df, x0=2, tol=1e-4)
print("Approximate root:", root)

# Plot
x = np.linspace(-5, 5, 400)
y = f(x)

plt.axhline(0, color='black', linewidth=1)
plt.plot(x, y, label="f(x)")
plt.scatter(roots, [f(r) for r in roots], color="red", label="Approximations")
plt.axvline(root, color="green", linestyle="--", label=f"Root ≈ {root:.4f}")
plt.legend()
plt.title("Newton-Raphson Method")
plt.show()
```
