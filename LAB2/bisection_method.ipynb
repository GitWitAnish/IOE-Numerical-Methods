# Bisection method

## Algorithm: Bisection Method

1. Define the function `f(x)` whose root is to be found.

2. Choose two initial guesses `a` and `b` such that `f(a) * f(b) < 0`  
   (they have opposite signs, so the root lies in `[a, b]`).

3. Set a tolerance value $tol$ (the acceptable error).

4. Create an empty list `roots` to store midpoint values for plotting.

5. Repeat while $(b - a) / 2 > tol$:
   - Compute midpoint:  
     $$m = \frac{a + b}{2}$$
   - Append $m$ to the list `roots`.  
   - If $|f(m)| < tol$, stop the loop (root found).  
   - Else if $f(a) \cdot f(m) < 0$, update $b = m$.  
   - Otherwise, update $a = m$.  

6. After the loop ends, approximate the root as:  
   $$\text{root} \approx \frac{a + b}{2}$$

7. Plot:  
   - The curve of `f(x)`  
   - Red dots at each midpoint stored in `roots`  
   - A vertical green dashed line at the final root  

8. Print the final root value.  


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the function
def f(x):
    return x**3 - 5*x - 9   # Example function

# Bisection Method with Simple Plot
def bisection_plot(f, a, b, tol):
    if f(a) * f(b) > 0:
        print("Invalid interval! f(a) and f(b) must have opposite signs.")
        return None
    
    roots = []   # store midpoints
    while (b - a) / 2 > tol:
        m = (a + b) / 2
        roots.append(m)

        if abs(f(m)) < tol:
            break
        elif f(a) * f(m) < 0:
            b = m
        else:
            a = m

    root = (a + b) / 2

    # --- Plotting ---
    x_vals = np.linspace(a-1, b+1, 400)
    y_vals = f(x_vals)

    plt.figure(figsize=(7,5))
    plt.plot(x_vals, y_vals, label="f(x)")
    plt.axhline(0, color="black", linewidth=0.8)

    # Midpoints
    plt.plot(roots, [f(r) for r in roots], "ro", label="Midpoints")

    # Final root
    plt.axvline(root, color="green", linestyle="--", label=f"Root â‰ˆ {root:.4f}")

    plt.legend()
    plt.grid(True)
    plt.show()

    print(f"Approx Root = {root:.8f}")
    return root


# Example usage
root = bisection_plot(f, 2, 3, 0.0001)
```
