 # Secant Method
 

## Algorithm: Secant Method

1. Define the function `f(x)` whose root is to be found.

2. Choose two initial guesses `x0` and `x1`, and a tolerance value `ε`.

3. Repeat until `|x1 - x0| < ε`:
   - Compute next approximation:  
     $$x_{2} = x_{1} - \frac{f(x_{1}) \cdot (x_{1} - x_{0})}{f(x_{1}) - f(x_{0})}$$
   - Update values:  
     $$x_{0} = x_{1}, \quad x_{1} = x_{2}$$

4. After the loop ends, approximate the root as `x1`.

5. Plot:  
   - Curve of `f(x)`  
   - Red dots at each approximation (`x2` values)  
   - A vertical green dashed line at the final root  

6. Print the final root value.  


```python
import numpy as np
import matplotlib.pyplot as plt

# Define function
def f(x):
    return x**3 - 5*x - 9   # Example function

# Secant Method
def secant_method(f, x0, x1, tol):
    roots = []
    while abs(x1 - x0) > tol:
        x2 = x1 - f(x1) * (x1 - x0) / (f(x1) - f(x0))
        roots.append(x2)
        x0, x1 = x1, x2
    return x1, roots

# Example usage
root, roots = secant_method(f, 2, 3, 1e-4)
print("Approximate root:", root)

# Plot
x = np.linspace(-5, 5, 400)
y = f(x)

plt.axhline(0, color='black', linewidth=1)
plt.plot(x, y, label="f(x)")
plt.scatter(roots, [f(r) for r in roots], color="red", label="Approximations")
plt.axvline(root, color="green", linestyle="--", label=f"Root ≈ {root:.4f}")
plt.legend()
plt.title("Secant Method")
plt.show()
```
